````python
plot.xticks(np.arange(len(labels)), labels, fontsize = "x-large", rotation = -30, ha = "left")
plot.yticks(np.linspace(0.3, 0.6, 16), fontsize = "x-large")
plot.ylim(0.3, 0.6)
plot.xlabel("Available Game Attributes", fontsize = "xx-large")
plot.ylabel(
    "Accuracy of Tuple Inclusion Guessing in Sampled Set\n(num. Correct Positive Guesses / num. Positive Guesses)",
    fontsize = "xx-large"
)

myColors = [
    '#' + format(h, "02x") + "3232"
    for h in (
        np.arange(accuracies.shape[0]) + accuracies.shape[0]
    ) * 255 // (2 * accuracies.shape[0])
]

#betas = np.linspace(0.05, 0.5, 10)
for i in range(0, accuracies.shape[0]):
    plot.plot(
        (0, accuracies.shape[1] - 1), (accuracies[i][0], accuracies[i][0]),
        linewidth = 0.75, linestyle = "--", color = myColors[i],
        zorder = i
    )
    plot.plot(
        accuracies[i],
        #label = "$\\beta =$ " + format(betas[i], " 4.2f"),
        linewidth = 1.5, color = myColors[i],
        linestyle = "--", zorder = i
    )
    plot.scatter( #bar(
        np.arange(accuracies.shape[1]), accuracies[i], 256.,
        label = "$\\beta =$ " + format(betas[i], " 4.2f"),
        color = myColors[i], marker = "X", zorder = i
    )
    
plot.legend(fontsize = "x-large")
plot.grid()
plot.title("Scammer Results", fontsize = "xx-large")
plot.savefig(
    "bigSimGenericizedWithKPlotScammer.png",
    facecolor = "white", bbox_inches = "tight"
)

plot.show()
````

````python
a1 = plot.axes(projection = "3d", azim = -90, elev = 7.5)
a1.set_xticks(
    np.arange(len(labels)), labels, fontsize = "x-large",
    rotation = -30, ha = "left"
)
a1.set_zticks(
    np.linspace(0., 0.3, 16), labels = [
        format(t, "04.2f") for t in np.linspace(0.3, 0.6, 16)
    ], fontsize = "x-large"
)
#a1.set_zlim(0.3, 0.6)
a1.set_xlabel("Available Game Attributes", labelpad = 100., fontsize = "xx-large")
a1.set_zlabel(
    "Accuracy of Tuple Inclusion Guessing in Sampled Set\n(num. Correct Positive Guesses / num. Positive Guesses)",
    labelpad = 50., fontsize = "xx-large"
)

myColors = [
    '#' + format(h, "02x") + "3232"
    for h in (
        np.arange(accuracies.shape[0]) + accuracies.shape[0]
    ) * 255 // (2 * accuracies.shape[0])
]

for i in range(0, accuracies.shape[0]):
    p = a1.bar3d(
        np.arange(accuracies.shape[1]), # x-axis positions (length)
        np.zeros(accuracies.shape[1]) + i, # y-axis (really z-axis) positions (depth)
        np.zeros(accuracies.shape[1]), # z-axis (really y-axis) positions (height)
        1., 1., accuracies[i] - 0.3,
        label = "$\\beta =$ " + format(betas[i], " 4.2f"), color = myColors[i],
        edgecolor = "black"
    )
    p._facecolors2d = p._facecolors # They still haven't fixed the bug with legend()
    p._edgecolors2d = p._edgecolors # So I have to put the colors here manually >:^(
    
a1.legend(loc = "upper left", fontsize = "x-large")
#a1.legend()
a1.grid()
a1.set_title("Scammer Results", fontsize = "xx-large")
plot.savefig(
    "bigSimGenericizedWithKPlotScammer3D.png",
    facecolor = "white", bbox_inches = "tight"
)

plot.show()
````
